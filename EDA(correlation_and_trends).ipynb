```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a simple dataset
data = {
    'Hours Studied': [1, 2, 3, 4, 5, 6],
    'Exam Score': [50, 55, 60, 70, 75, 80 ]
}

# Create a DataFrame
df = pd.DataFrame(data)

```

```python
df
```

```python
# Calculate the Pearson correlation coefficient
correlation = df['Hours Studied'].corr(df['Exam Score'])    
#the ".corr()" function is used to cal the strength of correlation between two variables
```

#### strength of relationship

```python
correlation
```

#### Direction of relationship

```python
#import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = df['Hours Studied'], y = df['Exam Score'], c = 'r', marker = "x")     #code to plot graph
plt.xlabel("Hours Studied")     #code to add tag to x axis
plt.ylabel("Exam Score")        #code to add tag to y axis
plt.title("Direction of graph")

plt.show()   
```

#### from the graph, we observed the direction of the relatioship (PROGRESSIVE "positive")

## Seasonal Trend

```python
import pandas as pd
```

```python
data_season = pd.read_csv("data_for_season.csv")
data_season  = data_season.head(12)
```

```python
#import visualization tools
import matplotlib.pyplot as plt
```

```python
x = data_season["Date"]
y = data_season["Sales"]
plt.figure(figsize=(15,8))       #to determine the size of the graph
plt.plot(x, y, linestyle = '-' )   #to plot the graph and select line type
plt.title("Sale seasonality Graph")  #graph title
plt.xlabel("Date")    #date tag
plt.ylabel("Sales")   #sales tag
plt.xticks(rotation = 45)   #to change the position or form of text for date tag
plt.legend()
plt.show()
```

## Cyclical Trend

```python
data_cycle = pd.read_csv("data_for_cycle.csv")
data_cycle = data_cycle.head(12)
data_cycle.columns
```

```python
x1 = data_cycle["Date"]
y1 = data_cycle["Revenue"]
plt.plot(x1, y1)
plt.title("Revenue by cyclical trend Graph")  #graph title
plt.xlabel("Date")    #date tag
plt.ylabel("Revenue")   #Revenue tag
plt.xticks(rotation = 45)   #to change the position or form of text for date tag
plt.legend()
plt.show()
```

## Structural Trend

```python
data_structure = pd.read_csv("data_for_structural_break.csv")
data_structure.head()
```

```python
x2 = data_structure["Date"]
y2 = data_structure["Revenue"]
plt.plot(x2, y2, linestyle = '-', label = "Revenue Trend")
plt.title("Revenue by Structural trend Graph")  #graph title
plt.axvline(data_structure["Date"][49], linestyle = '--', color = 'red')   #this marks a line on the graph. we placed the line at the revenu drop using indexing
plt.xlabel("Date")    #date tag
plt.ylabel("Revenue")   #Revenue tag
plt.xticks(rotation = 45)   #to change the position or form of text for date tag
plt.legend()
plt.show()
```
